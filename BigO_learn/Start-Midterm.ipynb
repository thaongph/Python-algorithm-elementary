```python
# create matrix:
n = int(input())
mat = []
for a in range(n):
    lst = list(map(int, input().split()))
    mat.append(lst)

# check max of row:
def check_row(mat, n, row, col):
    for j in range(n):
        if mat[row][j] > mat[row][col]:
            return False
    return True

def check_col(mat, n, row, col):
    for i in range(n):
        if mat[i][col] > mat[row][col]:
            return False
    return True
def check_parallel_diagonal(mat, n, row, col):
    for k in range(max(-row, -col), min(n-row-1, n-col-1)+1):
        if mat[row+k][col+k] > mat[row][col]:
            return False
    return True

def check_parallel_2nddiagonal(mat, n, row, col):
    for k in range(max(-row, -n+col+1), min(n-row-1, col)+1):
        if mat[row+k][col-k] > mat[row][col]:
            return False
    return True

def Queen(mat, n, row, col):
    if check_row(mat, n, row, col) and check_col(mat, n, row, col) and check_parallel_diagonal(mat, n, row, col) and check_parallel_2nddiagonal(mat, n, row, col):
        return True
    return False

count_queen = 0
for i in range(n):
    for j in range(n):
        if Queen(mat, n, i, j):
            count_queen +=1

print(count_queen)
```

### Lesson 3: Loop statement

Codeforces - A. Vanya and Cubes
https://codeforces.com/contest/492/problem/A

```python
cubes = int(input())
x = 0
used = 0
level = 1
while level > 0:
    x += level
    used += x
    if used == cubes:
        break
    elif used > cubes:
        level = level - 1
        break
    level +=1

print(level)
```
