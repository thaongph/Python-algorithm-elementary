```python
# Find greatest common division:
def gcd(a, b):
    while b != 0:
        r = a%b
        a = b
        b = r
    return a

# class Fraction and its methods:
class Fraction:
    # constructor:
    def __init__(self, num, denom):
        self.num = num
        self.denom = denom
    # method to sum up the fraction:
    def SumFra(self, other):
        a = self.num * other.denom + other.num * self.denom
        b = self.denom  * other.denom
        return Fraction(a,b)
    # method to reduce the result:
    def Reduce(self):
        c = gcd(self.num, self.denom)
        self.num //= c
        self.denom //= c

# Go with the input:
x, y = map(int, input().split())
a = Fraction(x,y)

x, y = map(int, input().split())
b = Fraction(x,y)

c = a.SumFra(b) # now, a and b can call methods created in class Fraction.
c.Reduce()

print(c.num, c.denom)
```

```python
class stu:
    def __init__(self, name, math, lit):
        self.name = str(name)
        self.math = float(math)
        self.lit = float(lit)
    def check_stu(self):
        if (self.math+self.lit)>=9:
            return True
        return False

n = int(input())
count_stu = 0
for i in range(n):
    name = input()
    math, lit = map(float, input().split())
    student = stu(name, math, lit)
    if student.check_stu():
        count_stu += 1
print(count_stu)
```

```python
class Point:
    def __init__(self, x, y):
        self.x =x
        self.y = y

    def distance(self, other ):
        return ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
    
    def __str__(self):
        return '{:.2f} {:.2f}'.format(self.x, self.y)
    
n = int(input())
res = 0
for i in range(n):
    x, y = map(int, input().split())
    A = Point(x, y)
    x, y = map(int, input().split())
    B = Point(x, y)
    x, y = map(int, input().split())
    C = Point(x, y)
    a = B.distance(C)
    b = A.distance(C)
    c = B.distance(A)
    p = (a+b+c)/2
    area = (p*(p-a)*(p-b)*(p-c))**0.5
    res += area

print("%.2f" % (res))
```

```python
class ribbon:
    def __init__(self):
        self.sum_length = 0
        self.count = 0
    def add_ribbon(self, length):
        self.sum_length += length
        self.count += 1

# Assign the integers from 1-1000 to class ribbon:
colors = []
for i in range(1001):
    colors.append(ribbon())

# Work with the input data
n = int(input())
d = {}
for i in range(n):
    p, l = map(int, input().split())
    colors[p].add_ribbon(l)
    d[p] = f"{colors[p].sum_length} {colors[p].count}"

print(len(d))

sorted_d = dict(sorted(d.items()))
for color, value in sorted_d.items():
    print(f"{color} {value}")
```

```python
# Find greatest common division:
def gcd(a, b):
    while b != 0:
        r = a%b
        a = b
        b = r
    return a

class Fraction: 
    def __init__(self, a, b):
        self.num = a
        self.denom = b
    def substract(self, other):
        return (other.num * self.denom - other.denom * self.num)
    def reduce(self):
        c = gcd(self.num, self.denom)
        self.num //= c
        self.denom //= c

n = int(input())
largest_fr = Fraction(0, 1001)
for i in range(n):
    x, y = map(int, input().split())
    fr = Fraction(x, y)
    if fr.substract(largest_fr) <= 0:
        largest_fr = fr

largest_fr.reduce()
print(largest_fr.num, largest_fr.denom)
```
