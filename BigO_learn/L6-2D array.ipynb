```python
# create matrix:
n = int(input())
mat = []
for a in range(n):
    lst = list(map(int, input().split()))
    mat.append(lst)

# check max of row:
def check_row(mat, n, row, col):
    for j in range(n):
        if mat[row][j] > mat[row][col]:
            return False
    return True

def check_col(mat, n, row, col):
    for i in range(n):
        if mat[i][col] > mat[row][col]:
            return False
    return True
def check_parallel_diagonal(mat, n, row, col):
    for k in range(max(-row, -col), min(n-row-1, n-col-1)+1):
        if mat[row+k][col+k] > mat[row][col]:
            return False
    return True

def check_parallel_2nddiagonal(mat, n, row, col):
    for k in range(max(-row, -n+col+1), min(n-row-1, col)+1):
        if mat[row+k][col-k] > mat[row][col]:
            return False
    return True

def Queen(mat, n, row, col):
    if check_row(mat, n, row, col) and check_col(mat, n, row, col) and check_parallel_diagonal(mat, n, row, col) and check_parallel_2nddiagonal(mat, n, row, col):
        return True
    return False

count_queen = 0
for i in range(n):
    for j in range(n):
        if Queen(mat, n, i, j):
            count_queen +=1

print(count_queen)
```

```python
# create matrix:
m, n = map(int, input().split())
mat = []
for a in range(n):
    lst = list(map(int, input().split()))
    mat.append(lst)

# find saddle value:
saddle =0
for i in range(m):
    for j in range(n):
        max_row = max(mat[i])
        min_col = min(row[j] for row in mat)
        if max_row == min_col:
            saddle +=1
            
print(saddle)
```
