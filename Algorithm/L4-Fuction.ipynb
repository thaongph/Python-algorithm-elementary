Codeforces - A. System of Equations
https://codeforces.com/problemset/problem/214/A

```python
import math
n, m = map(int, input().split())
pair = 0
for b in range(0,n+1):
    x = (n-b)**0.5
    if x % 1 == 0:
        a = x
        if (a + b**2) == m:
            pair += 1

print(pair)
```

Codeforces - A. Life without Zeros
https://codeforces.com/contest/75/problem/A

```python
a = int(input())
b = int(input())
c = a+b
def no_zero(x):
    lst = [i for i in str(x).split('0')]
    new_x = int(''.join(lst))
    return new_x
if no_zero(a) + no_zero(b) == no_zero(c):
    print('YES')
else: 
    print('NO')
```

```python
n = input()

def isPalindrome(k):
    lst = list(k)
    a = len(lst) // 2
    flag = True
    for i in range(0, a):
        if int(lst[i]) == int(lst[len(lst)-1-i]):
            flag = True
        else:
            flag = False
            break
    return flag

if isPalindrome(n) == True: 
    print('YES')
else:
    print('NO')
```
