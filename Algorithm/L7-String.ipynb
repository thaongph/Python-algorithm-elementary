A. Shortest path of the king - 
https://codeforces.com/problemset/problem/3/A

```python
lst = input().split()
revert = []
for word in lst[::-1]:
    revert.append(word)

str1 = ' '.join(x for x in revert)
print(str1)
```

```python
lst_str = input().split('. ')
lst_cptl = [lst_str[0]]
lst_str.pop(0)
for word in lst_str:
    if len(word) > 1:
        x = word[0].upper() + word[1:]
    else:
        x = word.upper()
    lst_cptl.append(x)
new_str = '. '.join(lst_cptl)
print(new_str)
```

```python
n = int(input())
for i in range(n):
    lst_str = list(input().upper())
    lst_str.sort()
    d = {}
    for cha in lst_str:
        d[cha] = d.get(cha, 0) + 1
    for key, value in d.items():
        if value == min(d.values()):
            print(key)
            break
```
