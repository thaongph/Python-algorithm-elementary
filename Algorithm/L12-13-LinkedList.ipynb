```python
class Node:
    def __init__(self, x):
        self.data = x
        self.next = None

class linkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    def insertTail(self, x):
        if self.head == None:
            p = Node(x)
            self.head = p
            self.tail = p
        else:
            p = Node(x)
            self.tail.next = p
            self.tail = p
    def removeHead(self):
        if self.head ==None:
            pass
        else:
            self.head = self.head.next

lst = linkedList()
n = int(input())
for i in range(n):
    inp = input()
    if len(inp) < 3:
      lst.removeHead()
    else:
        tempo_lst = [x for x in inp.split()]
        lst.insertTail(tempo_lst[1])

cur = lst.head
while cur != None:
    print(cur.data, end = ' ')
    cur = cur.next
```

```python
class Room:
    def __init__(self, id, price, available):
        self.id = str(id)
        self.price = int(price)
        self.available = int(available)

class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None

class linkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    def insertHead(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = self.tail = new_node
        else:
            self.head.prev = new_node
            self.head = new_node
    def emptyRoom(self):
        cur = self.tail
        while cur != None:
            if cur.data.available == 0:
                print(f"{cur.data.id} {cur.data.price}")
            cur = cur.prev
            
n = int(input())
lst = linkedList()
for i in range(n):
    x, y, z = input().split()
    p = Room(x, y, z)
    lst.insertHead(p)
    
lst.emptyRoom()
```
