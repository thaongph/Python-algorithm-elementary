```python
def factorial(x):
    if x<2:
        return 1
    return x* factorial(x-1)
n=int(input())
print(factorial(n))
```

```python
def odd_divisor(x):
    if x%2 ==1:
        return x
    else:
        x = int(x/2)
        return odd_divisor(x)

n = int(input())
print(odd_divisor(n))
```

```python
def count_num(x,y,num):
    if abs(x)//y ==0:
        return num
    else:
        num += 1
        y *= 10
        return count_num(x, y, num)
n = int(input())
print(count_num(n, 10, 1))
```

```python
import sys
sys.setrecursionlimit(10**5)

def isPrime(a):
    if a < 2:
        return False
    for i in range(2, int(a**0.5) +1):
        if a % i == 0:
            return False
    return True

def firstPrime(lst, index, n):
    if index == n:
        return -1
    if isPrime(lst[index]) == True:
        return index
    else:
        return firstPrime(lst, index+1, n)
    
n = int(input())
lst_num = list(map(int, input().split()))
index = 0
print(firstPrime(lst_num, index, n))
```

```python
import sys
sys.setrecursionlimit(10**5)

def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a%b)

a, b = map(int, input().split())
print(gcd(a, b))
```
