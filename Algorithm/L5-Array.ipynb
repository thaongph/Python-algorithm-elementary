```python
def isPrime(x):
    flag = True
    if x < 2:
        flag = False
    elif x> 2:
        for y in range(2,x):
            if x % y == 0:
                flag = False 
                break
    return flag

n = int(input())
ai = input()
num_lst = [int(i) for i in ai.split()[0:n]]
prime = 0
for num in num_lst:
    if isPrime(num) == True:
            prime +=1

print(prime)
```

```python
n = int(input())
str_h = input().split()
lst_h = [int(x) for x in str_h[0:n]]
min_h = min(lst_h)
E = 0
for h in lst_h:
    if h > min_h:
        E += h - min_h
print(E)
```
